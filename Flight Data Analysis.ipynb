## Q.2. Show Bar Graphs representing the Departure Time & Arrival Time.

```python
data.head()
```

```python
#showing the departure time for the flights

data['departure_time'].value_counts()
```

```python
#showing the arrival time for the flights

data['arrival_time'].value_counts()
```

```python
#showing the departure time and arrival time for the flight with their counts

plt.figure(figsize = (16,4))

# Departure subplot
plt.subplot(1, 2, 1)
departure_counts = data['departure_time'].value_counts()
plt.bar(
    departure_counts.index, 
    departure_counts.values, 
    color=plt.cm.Blues(departure_counts / departure_counts.max())
)
plt.title("Departure Flight Frequencies")
plt.xlabel("Departure Time")
plt.ylabel("Frequency")

# Arrival subplot
plt.subplot(1, 2, 2)
arrival_counts = data['arrival_time'].value_counts()
plt.bar(
    arrival_counts.index, 
    arrival_counts.values, 
    color=plt.cm.Greens(arrival_counts / arrival_counts.max())
)
plt.title("Arrival Flight Frequencies")
plt.xlabel("Arrival Time")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

## Q.5. Does ticket price change based on the departure time and arrival time?


```python
data.head()
```

```python
#check avg ticket price for each departure time

data.groupby('departure_time')['price'].mean()
```

```python
#check avg ticket price for each arrival time

data.groupby('arrival_time')['price'].mean()
```

```python
plt.figure(figsize=(16, 6))

#Departure Time subplot
plt.subplot(1, 2, 1)
# Aggregate mean price by departure_time and sort descending
departure_avg = data.groupby('departure_time')['price'].mean().reset_index()
departure_avg = departure_avg.sort_values(by='price', ascending=False)

sns.barplot(
    x='departure_time',
    y='price',
    data=departure_avg,
    palette=sns.color_palette("Blues_d", len(departure_avg))
)
plt.title("Average Ticket Price by Departure Time")
plt.xlabel("Departure Time")
plt.ylabel("Average Price")
plt.xticks(rotation=45)

#Arrival Time subplot 
plt.subplot(1, 2, 2)
# Aggregate mean price by arrival_time and sort descending
arrival_avg = data.groupby('arrival_time')['price'].mean().reset_index()
arrival_avg = arrival_avg.sort_values(by='price', ascending=False)

sns.barplot(
    x='arrival_time',
    y='price',
    data=arrival_avg,
    palette=sns.color_palette("Greens_d", len(arrival_avg))
)
plt.title("Average Ticket Price by Arrival Time")
plt.xlabel("Arrival Time")
plt.ylabel("Average Price")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
sns.relplot( x='arrival_time', y='price', data=data, col='departure_time', kind='line')
plt.show
```
